```python
import os
captSch_files=os.listdir ('../data/captain_schedule/')[:-1]
df=pd.DataFrame()
for captSch_file in captSch_files:
    print(captSch_file)
    busDriven=pd.read_csv('../data/captain_schedule/'+captSch_file)[['cs_busno','depot_id','route_id','cs_date']]
    df=pd.concat([df,busDriven])

df=df.rename(columns={'cs_busno':'bus_no'})

df.to_csv('../data/processed/busList.csv',index=False)
```

```python
import os
td_files=os.listdir ('../data/trip_departure/')[1:]
df=pd.DataFrame()
for td_file in td_files:
    print(td_file)
    td=pd.read_csv('../data/trip_departure/'+td_file)[['ope_date','depot_id','line_id','bus_no']]
    df=pd.concat([df,td])
    df=df.drop_duplicates()

df.to_csv('../data/processed/busList.csv',index=False)
```

```python
accident=pd.read_csv('../data/accident.csv')[['bus_no','capt_id','bcc_acc_id','acc_date','depot_id','route_id','bus_damage','eng_find']]
accident['bus_damage']=accident['bus_damage'].str.upper()
accident['eng_find']=accident['eng_find'].str.upper()
accident['label_1']='accident'
accident
```

```python
eng_bus_repair=pd.read_csv('../data/eng_bus_repair.csv')[['br_id','depot_id','br_busno','br_repair_status','br_entry_status','br_date','br_date_entry','br_target_completed']]
eng_bus_repair=eng_bus_repair.rename(columns={'br_busno':'bus_no'})

eng_bus_repair['br_target_completed']=np.where(eng_bus_repair['br_target_completed'].str.contains('0202-03-20'),'2020-03-20',eng_bus_repair['br_target_completed']) # data clean

eng_bus_repair['br_ttl_days']=(pd.to_datetime(eng_bus_repair['br_target_completed'])-pd.to_datetime(eng_bus_repair['br_date_entry'])).dt.days
eng_bus_repair['br_entry_status']=eng_bus_repair['br_entry_status'].str.upper()

eng_bus_repair=eng_bus_repair.drop_duplicates()

eng_bus_repair['label_2']='engine repair'
eng_bus_repair
```

```python
rtd=pd.read_csv('../data/rtd.csv')[['rtd_id','rtd_date','depot_id','route_id','bus_no','bus_model_id','driver_id','rtd_cause','cause_remark','fc_remark']] #to hi lite descriptive stat for repair cause , need an etl to standardize , rtd --> return to depot?
rtd['cause_remark']=rtd['cause_remark'].str.upper()
rtd['fc_remark']=rtd['fc_remark'].str.upper()
rtd['label_3']='return to depot'
rtd
```

```python
historical_issues=pd.merge(eng_bus_repair,rtd,how='left',left_on=['br_date_entry','depot_id','bus_no'],right_on=['rtd_date','depot_id','bus_no'] )
historical_issues
```

```python
rtd[['br_repair_status','br_entry_status','br_date','br_date_entry','br_target_completed','br_ttl_days']]=np.NaN
rtd
```

```python
historical_issues=pd.concat([historical_issues,rtd]).drop_duplicates()
historical_issues['date_issues']=np.nan
historical_issues['date_issues']=historical_issues['br_date_entry']
historical_issues['date_issues']=np.where(historical_issues['date_issues'].isnull(),historical_issues['rtd_date'],historical_issues['date_issues'])
historical_issues
```

```python
historical_issues=pd.merge(historical_issues,accident,how='left',left_on=['depot_id','bus_no','date_issues','route_id'],right_on=['depot_id','bus_no','acc_date','route_id'])
historical_issues=historical_issues.drop_duplicates()
historical_issues
```

```python
depot=pd.read_csv('../data/depot.csv')
depot['state']=np.where(depot['state_id']==1,'Selangor/Kuala Lumpur',np.nan)
depot['state']=np.where(depot['state_id']==8,'Pahang',depot['state'])
depot['state']=np.where(depot['state_id']==5,'Penang',depot['state'])

depot=depot[['depot_id','depot_name','state','zone','code']]
```

```python
route=pd.read_csv('../data/route.csv')[['depot_id','route_id','route_desc']]
route
```

```python
route_busstop=pd.read_csv('../data/route_busstop.csv')[['route_id','bus_stop_id','distance']]
route_busstop
```

```python
bus_stop_list=pd.read_csv('../data/bus_stop_list.csv')[['bus_stop_id','stop_name','street_name','latitude','longitude']]
bus_stop_list
```

### Info Depot

```python
InfoDEPOT=pd.merge(depot,route,how='left',on='depot_id')
InfoDEPOT=pd.merge(InfoDEPOT,route_busstop,how='left',on='route_id')
InfoDEPOT=pd.merge(InfoDEPOT,bus_stop_list,how='left',on='bus_stop_id')
InfoDEPOT.to_csv('../data/processed/InfoDEPOT.csv',index=False)
```

### Info Bas

```python
InfoBAS=pd.read_csv('../data/processed/busList.csv') #this data show that buses is moving from one depot to another and also embarked at many diffrent route id
InfoBAS=pd.merge(InfoBAS,historical_issues,how='left',left_on=['bus_no','depot_id','ope_date'],right_on=['bus_no','depot_id','date_issues'])
InfoBAS['label']=InfoBAS['label_1']
InfoBAS['label']=np.where(InfoBAS['label'].isnull(),InfoBAS['label_2'],InfoBAS['label'])
InfoBAS['label']=np.where(InfoBAS['label'].isnull(),InfoBAS['label_3'],InfoBAS['label'])
InfoBAS['label']=np.where(InfoBAS['label'].isnull(),'normal trip',InfoBAS['label'])
InfoBAS=InfoBAS.drop_duplicates()

InfoBAS.to_csv('../data/processed/InfoBAS.csv',index=False)
```
