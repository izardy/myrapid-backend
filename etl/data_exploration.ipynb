```python
import warnings
warnings.filterwarnings('ignore')

days=list(range(1,32,1))
months=list(range(1,2,1))
years=list(range(2022,2023,1))
bus_ids=trip_departure_1_2022['bus_no'].drop_duplicates().to_list()
df=pd.DataFrame()

for bus_id in bus_ids:
    data=trip_departure_1_2022[trip_departure_1_2022['bus_no']==bus_id]
    data['dt_gps']=pd.to_datetime(data['dt_gps'])
    data=data.sort_values('dt_gps')

    for year in years:
        data['dt_gps'].dt.year==year

        for month in months:
            data['dt_gps'].dt.month==month

            for day in days:
                data['dt_gps'].dt.day==day

                data['diff_minutes']=pd.to_datetime(data['dt_gps']).diff()/np.timedelta64(1, 'm')
                data['diff_label']=np.where((data['dir']=='arrive')&(data['direction']=='DIR2_arrive'),'to point a',np.nan)
                data['diff_label']=np.where((data['dir']=='arrive')&(data['direction']=='DIR1_arrive'),'from point a',data['diff_label'])
                data['diff_label']=np.where((data['dir']=='depart')&(data['direction']=='DIR1_depart'),'idle mode',data['diff_label'])
                data['diff_label']=np.where((data['dir']=='depart')&(data['direction']=='DIR2_depart'),'idle mode',data['diff_label'])

                data_1=data.groupby(['ope_date','bus_no'])[['td_id']].count().reset_index().rename(columns={'td_id':'ttl_td_id'}) # total trips on date
                data_2=data.groupby(['ope_date','bus_no'])[['captain_id']].nunique().reset_index().rename(columns={'captain_id':'ttl_captain_id'}) # total captains on duty on date
                data_3=data.groupby(['ope_date','bus_no'])[['point_a']].nunique().reset_index().rename(columns={'point_a':'ttl_point_a'}) # total point a of trip on date
                data_4=data.groupby(['ope_date','bus_no'])[['point_b']].nunique().reset_index().rename(columns={'point_b':'ttl_point_b'}) # total point b of trip on date

                data_5=data[data['diff_label']=='from point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'max'}).reset_index().\
                    rename(columns={'diff_minutes':'max_diff_minutes_from_point_a'}).drop(columns=['diff_label'])
                data_6=data[data['diff_label']=='idle mode'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'max'}).reset_index().\
                    rename(columns={'diff_minutes':'max_diff_minutes_idle_mode'}).drop(columns=['diff_label'])
                data_7=data[data['diff_label']=='to point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'max'}).reset_index().\
                    rename(columns={'diff_minutes':'max_diff_minutes_to_point_a'}).drop(columns=['diff_label'])

                data_8=data[data['diff_label']=='from point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'min'}).reset_index().\
                    rename(columns={'diff_minutes':'min_diff_minutes_from_point_a'}).drop(columns=['diff_label'])
                data_9=data[data['diff_label']=='idle mode'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'min'}).reset_index().\
                    rename(columns={'diff_minutes':'min_diff_minutes_idle_mode'}).drop(columns=['diff_label'])
                data_10=data[data['diff_label']=='to point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'min'}).reset_index().\
                    rename(columns={'diff_minutes':'min_diff_minutes_to_point_a'}).drop(columns=['diff_label'])

                data_11=data[data['diff_label']=='from point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'mean'}).reset_index().\
                    rename(columns={'diff_minutes':'avg_diff_minutes_from_point_a'}).drop(columns=['diff_label'])
                data_12=data[data['diff_label']=='idle mode'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'mean'}).reset_index().\
                    rename(columns={'diff_minutes':'avg_diff_minutes_idle_mode'}).drop(columns=['diff_label'])
                data_13=data[data['diff_label']=='to point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'mean'}).reset_index().\
                    rename(columns={'diff_minutes':'avg_diff_minutes_to_point_a'}).drop(columns=['diff_label'])

                data_14=data[data['diff_label']=='from point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'median'}).reset_index().\
                    rename(columns={'diff_minutes':'med_diff_minutes_from_point_a'}).drop(columns=['diff_label'])
                data_15=data[data['diff_label']=='idle mode'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'median'}).reset_index().\
                    rename(columns={'diff_minutes':'med_diff_minutes_idle_mode'}).drop(columns=['diff_label'])
                data_16=data[data['diff_label']=='to point a'].groupby(['ope_date','bus_no','diff_label']).agg({'diff_minutes':'median'}).reset_index().\
                    rename(columns={'diff_minutes':'med_diff_minutes_to_point_a'}).drop(columns=['diff_label'])

                data_17=data.groupby(['ope_date','bus_no']).agg({'speed':'max'}).reset_index().rename(columns={'speed':'max_speed'})
                data_18=data.groupby(['ope_date','bus_no']).agg({'speed':'min'}).reset_index().rename(columns={'speed':'min_speed'})
                data_19=data.groupby(['ope_date','bus_no']).agg({'speed':'mean'}).reset_index().rename(columns={'speed':'avg_speed'})
                data_20=data.groupby(['ope_date','bus_no']).agg({'speed':'median'}).reset_index().rename(columns={'speed':'med_speed'})

                data_21=data.groupby(['ope_date','bus_no']).agg({'is_completed':'sum'}).reset_index().rename(columns={'is_completed':'ttl_td_completed'})

                data_int=pd.merge(data_1,data_2,how='left',on=['ope_date','bus_no'])

                data_int=pd.merge(data_int,data_3,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_4,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_5,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_6,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_7,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_8,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_9,how='left',on=['ope_date','bus_no'])

                data_int=pd.merge(data_int,data_10,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_11,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_12,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_13,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_14,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_15,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_16,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_17,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_18,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_19,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_20,how='left',on=['ope_date','bus_no'])
                data_int=pd.merge(data_int,data_21,how='left',on=['ope_date','bus_no'])

                df=pd.concat([df,data_int])

df
```
